## Background of the COVID-19 Pandemic

The novel coronavirus disease that began rapid spread in early 2020 has been referred to as "COVID-19". This disease has spread through many countries throughout the world and as of May 2022, has not been eradicated. [COVID-19 is generally spread](https://www.cdc.gov/coronavirus/2019-ncov/prevent-getting-sick/how-covid-spreads.html) through infected saliva or air molecules released by an individual that is infected, which is why many organizations worldwide endorse wearing a face covering to reduce community transmission.

Vaccine development began as soon as the virus was identified using new technologies in some cases and vaccines became available to the public in late 2020 and early 2021, depending on location. One of the new technologies in particular that was used was an mRNA vaccine which focused on targeting proteins to reduce the likelihood of infection. Some countries developed their own vaccines if possible, but others adopted the importing of vaccines from countries with vaccines that were developed.

## The Sum of Cases per Continent

```python
import pandas
```

```python
df = pandas.read_excel('s3://ia241-solderich/covid_data.xls')
```

```python
import as plt
```

```python
sum_cases_by_continent = df.groupby('continentExp').sum()['cases']
sum_cases_by_continent.nlargest(10).plot.bar()

plt.title ('Cases per Continent')
plt.xlabel ('Continent' )
plt.ylabel ('Cases per continent in millions')
plt. show ()
```

Since the transmission of COVID-19 is spread through the air, as previously discussed, it is important to gather where the highest number of cases are. Travel is much easier and quicker through a landmass than overseas, so to avoid further spread it would be wise to avoid moving to continents with a high number of cases. As can be seen when cases are graphed by continent, America has the highest number of cumulative cases, followed by Europe.

## Top 10 Countries with the highest Death Rates per capita

```python
death_per_capita_by_country = df.groupby('countriesAndTerritories').sum()['deaths']/df.groupby('countriesAndTerritories').mean()['popData2019']
death_per_capita_by_country.nlargest(10).plot.bar()
plt.title ('Highest COVID death rate per capita')
plt.xlabel ('Country' )
plt.ylabel ('Deaths per capita')
plt. show ()
```

The countries graphed are those with the highest death rate per capita, the highest rate being Belgium. In many cases, the smaller the overall population, the more of an impact even a singular death makes on the deaths per capita. These graphs were calculated using the sum of deaths in each country divided by the average population per country. Taking the average population in the country is necessary since with this dataset the population is repeated multiple times.

## How did China's number of cases progress throughout the year?

```python
china_data = df.loc[df['countriesAndTerritories'] == 'China']
```

```python
china_cases = china_data.groupby('month').sum()['cases']
```

```python
china_cases.plot(x='month', y='cases')
plt.title ('China COVID Cases over time')
plt.xlabel ('Month' )
plt.ylabel ('Cases ')
```

The line chart for COVID cases in China shows the progression of China's cases throughout the 12 motnhs of the year. There is an apparent spike of cases in February (month 2) and the case count is generally low throughout the remainder of the year. While there is no direct reasoning for this, it is possible that it is due to February being one of the colder months in the northern hemisphere causing individuals to be indoors with others. The case count peaks at 70,000 in February.
